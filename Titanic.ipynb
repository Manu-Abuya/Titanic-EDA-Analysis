```python
import pandas as pd;
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv("titanic.csv")
```

```python
titanic_data.info()
```

```python
titanic_data.head()
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data
```

```python
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
```

```python
titanic_data['Fare'].fillna(titanic_data['Fare'].median(), inplace=True)
```

```python
titanic_data['Cabin'].fillna('Unknown', inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.head()
```

```python
# Passenger demographics
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

# Counterplot for 'Sex'
plt.subplot(1, 2, 1)
sns.countplot(x='Sex', data=titanic_data, palette='Set1')
plt.title('Passenger Gender Distribution')
```

```python
# countplot for 'Pclass'
plt.subplot(1, 2, 2)
sns.countplot(x='Pclass', data=titanic_data, palette='Set2')
plt.title('Passenger Class Distribution')

plt.show()
```

```python
# Analyze Survival Rates based on sex
plt.figure(figsize=(12, 6))

sns.countplot(x='Survived', hue='Sex', data=titanic_data, palette='pastel')
plt.title('Survival Count by Gender')

plt.show()
```

```python
# Visualize ticket class and cabin locations
plt.figure(figsize=(14, 6))

# Plot a countplot for 'Pclass' based on 'Survived'
plt.subplot(1, 2, 1)
sns.countplot(x='Pclass', hue='Survived', data=titanic_data, palette='dark')
plt.title('Survival Count by Passenger Class')

# Plot a heatmap for 'Cabin' data
plt.subplot(1, 2, 2)
cabin_survival = titanic_data[['Cabin', 'Survived']].dropna()
sns.heatmap(pd.crosstab(cabin_survival['Cabin'], cabin_survival['Survived']), cmap='coolwarm', annot=True, fmt='d')
plt.title('Survival Heatmap by Cabin')

plt.show()
```
